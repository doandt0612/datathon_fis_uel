```python
import pandas as pd
import numpy as np
import os
```

## Đường dẫn tương đối theo cấu trúc project


```python
current_dir = os.getcwd()
if os.path.exists(os.path.join('data', 'raw')):
    BASE_DIR = '.'
elif os.path.exists(os.path.join('..', 'data', 'raw')):
    BASE_DIR = '..'
else:
    raise FileNotFoundError("❌ Không tìm thấy thư mục data/raw.")
```

## Khởi tạo thư mục đích


```python
RAW_DIR = os.path.join(BASE_DIR, "data", "raw")
CLEAN_DIR = os.path.join(BASE_DIR, "data", "processed", "clean")

os.makedirs(CLEAN_DIR, exist_ok=True)
print(f"✅ Đã thiết lập xong đường dẫn. Dữ liệu sạch sẽ được lưu tại: {CLEAN_DIR}")
```

## 1/File customers

# 1. Load data


```python
df_cust = pd.read_csv(os.path.join(RAW_DIR, "customers.csv"))
df_cust.head()
```

## 1.1. Kiểm tra null và điền unknown vào ô trống

```python
df_cust['acquisition_channel'] = df_cust['acquisition_channel'].fillna('unknown')
df_cust.head()
```

## 2. Ép kiểu và trích xuất thời gian

```python

df_cust['signup_date'] = pd.to_datetime(df_cust['signup_date'])
df_cust['signup_year'] = df_cust['signup_date'].dt.year
df_cust['signup_month'] = df_cust['signup_date'].dt.month
df_cust.head()
```

## 3. Lưu file

```python
df_cust.to_csv(os.path.join(CLEAN_DIR, "customers_cleaned.csv"), index=False)
print(f"✅ Đã lưu customers_cleaned.csv với {len(df_cust)} dòng.")
```

1.Xử lý dữ liệu Khách hàng (customers_cleaned.csv)
- Làm sạch (Cleaning): Xử lý các giá trị khuyết (Missing Values) tại cột acquisition_channel (kênh thu hút) bằng cách điền nhãn 'unknown' thay vì xóa dòng. Việc này giúp bảo toàn nguyên vẹn số lượng khách hàng.

- Tạo đặc trưng (Feature Engineering): Trích xuất signup_year (Năm đăng ký) và signup_month (Tháng đăng ký) từ cột signup_date ban đầu.

- Ý nghĩa Nghiệp vụ: Cung cấp sẵn các chiều dữ liệu thời gian để team Business dễ dàng xây dựng bộ lọc (Slicers) và phân tích xu hướng tăng trưởng tập khách hàng mới trên các Dashboard.

## 2/File geography

## 1. Load data

```python
df_geo = pd.read_csv(os.path.join(RAW_DIR, "geography.csv"))
df_geo.head()
```

## 2. Xóa trùng lặp mã zip

2.Xử lý dữ liệu Địa lý (geography_cleaned.csv)
- Làm sạch (Cleaning): Chủ động kiểm tra và loại bỏ các dòng trùng lặp (Drop Duplicates) dựa trên khóa chính là mã bưu điện (zip).

- Ý nghĩa Nghiệp vụ: Đảm bảo tính duy nhất (Unique) của dữ liệu không gian. Đây là bước phòng thủ cực kỳ quan trọng để ngăn chặn lỗi "nổ dòng" (Data Explosion) làm sai lệch doanh thu khi thực hiện phép nối bảng (Left Join) với hệ thống bán hàng ở các bước sau.

## 3/File products

## 1. Load data

```python
df_prod = pd.read_csv(os.path.join(RAW_DIR, "products.csv"))
df_prod.head()
```

## 1.1. Xóa những sản phẩm bị thiếu thông tin về giá và sau đó in ra xem còn lại bao nhiêu sản phẩm đạt chuẩn.

```python
df_prod = df_prod.dropna(subset=['price', 'cogs']).copy()
print(f"Sản phẩm hợp lệ: {len(df_prod)}")
df_prod.head()
```

## 2. Tính Profit Margin

```python
df_prod['profit_margin'] = (df_prod['price'] - df_prod['cogs']) / df_prod['price']
df_prod.head()
```

## 3. Lưu file

```python
df_prod.to_csv(os.path.join(CLEAN_DIR, "products_cleaned.csv"), index=False)
print("✅ Đã lưu products_cleaned.csv")
```

3.Xử lý dữ liệu Sản phẩm (products_cleaned.csv)
- Làm sạch (Cleaning): Loại bỏ hoàn toàn các sản phẩm bị khuyết thông tin price (Giá bán) hoặc cogs (Giá vốn).

- Tạo đặc trưng (Feature Engineering): Xây dựng chỉ số profit_margin (Biên lợi nhuận) cho từng mặt hàng theo công thức: (Giá bán - Giá vốn) / Giá bán.

- Ý nghĩa Nghiệp vụ: Các sản phẩm mất dấu vết tài chính không có giá trị phân tích kinh tế nên bắt buộc phải bị loại trừ. Đặc trưng Biên lợi nhuận mới tạo là thước đo cốt lõi giúp phân loại sản phẩm thành nhóm "gà đẻ trứng vàng" hay nhóm "sản phẩm mồi".

## 4/ File promotions

## 1. Load data

```python
df_promo = pd.read_csv(os.path.join(RAW_DIR, "promotions.csv"))
df_promo.head()
```

## 1.1. Fill NaN bằng 'All'


```python
df_promo['applicable_category'] = df_promo['applicable_category'].fillna('All')
df_promo.isnull().sum()
```

```python
df_promo.head()
```

## 2. Tính số ngày diễn ra khuyến mãi


```python
df_promo['start_date'] = pd.to_datetime(df_promo['start_date'])
df_promo['end_date'] = pd.to_datetime(df_promo['end_date'])
df_promo['promo_duration_days'] = (df_promo['end_date'] - df_promo['start_date']).dt.days
df_promo.head()
```

## 3. Lưu file

```python
df_promo.to_csv(os.path.join(CLEAN_DIR, "promotions_cleaned.csv"), index=False)
print("✅ Đã lưu promotions_cleaned.csv")
```

4.Xử lý dữ liệu Khuyến mãi (promotions_cleaned.csv)
- Làm sạch (Cleaning): Gán nhãn 'All' cho các giá trị NaN tại cột applicable_category (Danh mục áp dụng).

- Tạo đặc trưng (Feature Engineering): Tính toán độ dài của chiến dịch promo_duration_days (Số ngày KM) bằng khoảng cách giữa ngày kết thúc và ngày bắt đầu.

- Ý nghĩa Nghiệp vụ: Việc gán 'All' phản ánh đúng thực tế: mã giảm giá không quy định danh mục nghĩa là áp dụng cho toàn sàn. Biến số thời lượng khuyến mãi là công cụ đắc lực để phân tích hiệu ứng FOMO (Sale chớp nhoáng) so với các chương trình duy trì dài hạn.

## 5/ File inventory

## 1. Load data

```python
df_inv = pd.read_csv(os.path.join(RAW_DIR, "inventory.csv"))
df_inv.head()
```

## 1.1. Chuyển đổi dữ liệu của cột snapshot_date từ dạng chữ (text/string) sang dạng thời gian chuẩn (datetime)

```python
df_inv['snapshot_date'] = pd.to_datetime(df_inv['snapshot_date'])
df_inv.head()
```

## 2. Tính vòng quay hàng tồn (Cộng 1e-5 để tránh lỗi ZeroDivisionError)

```python
df_inv['inventory_turnover'] = df_inv['units_sold'] / (df_inv['stock_on_hand'] + 1e-5)
df_inv[['product_id', 'stock_on_hand', 'units_sold', 'inventory_turnover']].head()
```

## 3. Lưu file

```python
df_inv.to_csv(os.path.join(CLEAN_DIR, "inventory_cleaned.csv"), index=False)
print("✅ Đã lưu inventory_cleaned.csv")
```

5.Xử lý dữ liệu Tồn kho (inventory_cleaned.csv)
- Làm sạch (Cleaning): Chuẩn hóa kiểu dữ liệu (Type Casting) cho cột snapshot_date về đúng định dạng Datetime.

- Tạo đặc trưng (Feature Engineering): Tính toán inventory_turnover (Vòng quay tồn kho) bằng cách lấy số lượng bán chia cho lượng tồn kho hiện tại (cộng thêm một hằng số cực nhỏ 1e-5 để chống lỗi chia cho 0).

- Ý nghĩa Nghiệp vụ: Cung cấp góc nhìn thực tế về tốc độ giải phóng hàng hóa. Chỉ số này giúp quản lý chuỗi cung ứng nhận diện ngay lập tức các mặt hàng đang bị ứ đọng vốn (Turnover cực thấp) hoặc đang có nguy cơ cháy kho (Turnover cực cao).

## 6/ File web_trafic

## 1. Load data

```python
df_web = pd.read_csv(os.path.join(RAW_DIR, "web_traffic.csv"))
df_web.head()
```

## 1.1. Chuyển đổi cột 'date' từ dạng chữ (text) sang dạng thời gian chuẩn (datetime)

```python
df_web['date'] = pd.to_datetime(df_web['date'])
df_web.head()
```

## 2. Phân rã dữ liệu chuỗi thời gian thành day_of_week và is_weekend (ngày nghỉ)

```python
df_web['day_of_week'] = df_web['date'].dt.day_name()
df_web['is_weekend'] = df_web['date'].dt.dayofweek.isin([5, 6]).astype(int)
df_web[['date', 'day_of_week', 'is_weekend']].head()
```

## 3. Lưu file

```python
df_web.to_csv(os.path.join(CLEAN_DIR, "web_traffic_cleaned.csv"), index=False)
print("✅ Đã lưu web_traffic_cleaned.csv")
```

6.Xử lý dữ liệu Lưu lượng Web (web_traffic_cleaned.csv)
- Làm sạch (Cleaning): Đưa toàn bộ cột thời gian truy cập date về chuẩn định dạng ngày tháng.

- Tạo đặc trưng (Feature Engineering): Bóc tách day_of_week (Thứ trong tuần) và thiết lập cờ nhị phân is_weekend (1 = Cuối tuần, 0 = Ngày thường).

- Ý nghĩa Nghiệp vụ: Trang bị sẵn các bộ đếm thời gian linh hoạt. Phân hệ Marketing có thể đối chiếu trực tiếp hành vi lướt web và tỷ lệ chuyển đổi (Conversion Rate) giữa ngày thường và ngày nghỉ để tối ưu hóa ngân sách chạy quảng cáo.
